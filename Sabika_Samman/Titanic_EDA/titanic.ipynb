```python
df['Survived'].value_counts().plot(kind='bar', color=['red','green'])
plt.title("Survived vs Not Survived")
plt.xlabel("0 = Not Survived, 1 = Survived")
plt.ylabel("Number of Passengers")
plt.show()
```

 ## Filtering 

```python
Survived_passengers = df[df['Survived'] == 1]
print("Total survivors:", len(Survived_passengers))
Survived_passengers.head()
```

```python
# Only female passengers who were in First Class
female_first_class = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
female_first_class.head(10)
```

```python
# Only male passengers who were in First Class
male_first_class = df[(df['Sex'] == 'male') & (df['Pclass'] == 1)]
male_first_class.head(10)
```

```python
seniors = df[df['Age'] > 60]
print(len(seniors))
```

## Simple Visualization (Matplotlib)

```python
df['Sex'].value_counts().plot(kind='bar', color=['steelblue','salmon'])
plt.title('Passenger Count by Sex')
plt.ylabel('Count')
plt.show()
```

```python
df.groupby('Pclass')['Survived'].mean().plot(kind='bar', color='seagreen')
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```
