## 1️⃣ Pandas Series

```python
import pandas as pd
import numpy as np

# A Series is like a 1-dimensional array with labels
marks = pd.Series([85, 92, 78, 90, 88], index=["Ali", "Sara", "Bilal", "Ayesha", "Usman"])
marks
```

```python
# Basic operations on a Series
print("Max:", marks.max())
print("Min:", marks.min())
print("Mean:", marks.mean())
print("Value at Sara:", marks["Sara"])
```

## 2️⃣ Creating a DataFrame

A DataFrame is the most important structure in Pandas — it is a table (rows x columns).

```python
data = {
    "Name": ["Ali", "Sara", "Bilal", "Ayesha", "Usman"],
    "Age": [23, 21, 25, 22, 24],
    "City": ["Rawalpindi", "Lahore", "Karachi", "Islamabad", "Multan"],
    "Marks": [85, 92, 78, 90, 88]
}

df = pd.DataFrame(data)
df
```

## 3️⃣ Reading / Writing CSV Files

```python
# Save DataFrame to a CSV file
df.to_csv("students.csv", index=False)

# Load it back from the CSV file
df = pd.read_csv("students.csv")
df
```

## 5️⃣ Selection & Indexing

`loc` = label-based selection
`iloc` = position/index-based selection

```python
# Selecting a single column
df["Name"]
```

```python
# Selecting multiple columns
df[["Name", "Marks"]]
```

```python
# loc -> label based (row label, column label)
df.loc[0]          # first row
```

```python
df.loc[0, "Name"]   # specific value
```

```python
# iloc -> position based
df.iloc[1]          # 2nd row (index 1)
```

```python
df.iloc[0:3, 0:2]   # first 3 rows, first 2 columns
```

## 6️⃣ Filtering Rows (Conditions)

```python
# Students who scored above 85
df[df["Marks"] > 85]
```

```python
# Multiple conditions (AND -> &, OR -> |)
df[(df["Marks"] > 80) & (df["Age"] < 24)]
```

```python
# Data for a specific city
df[df["City"] == "Lahore"]
```

## 7️⃣ Adding/Removing Columns

```python
# Adding a new column
df["Grade"] = ["A", "A+", "B+", "A", "A-"]
df
```

```python
# Creating a new column based on a calculation from an existing column
df["Marks_Percentage"] = (df["Marks"] / 100) * 100
df
```

```python
# Removing a column
df = df.drop(columns=["Marks_Percentage"])
df
```

## 8️⃣ Sorting

```python
# Sort by Marks in descending order
df.sort_values(by="Marks", ascending=False)
```

```python
# Sort by multiple columns
df.sort_values(by=["City", "Age"])
```

## 9️⃣ Handling Missing Values

```python
# Example dataset with a missing value
df2 = df.copy()
df2.loc[2, "Marks"] = np.nan
df2
```

```python
# Checking for missing values
df2.isnull().sum()
```

```python
# Filling a missing value (with the mean)
df2["Marks"] = df2["Marks"].fillna(df2["Marks"].mean())
df2
```

```python
# Dropping rows with missing values (if needed)
# df2.dropna()
```

## 🔟 GroupBy Basics

```python
# Average marks by city
df.groupby("City")["Marks"].mean()
```

```python
# Multiple aggregations
df.groupby("City")["Marks"].agg(["mean", "max", "min"])
```

## 1️⃣1️⃣ Apply Function

```python
# Applying a custom function to each row
def pass_fail(marks):
    return "Pass" if marks >= 80 else "Fail"

df["Result"] = df["Marks"].apply(pass_fail)
df
```

## 1️⃣2️⃣ Basic Statistics

```python
print("Average Marks:", df["Marks"].mean())
print("Highest Marks:", df["Marks"].max())
print("Lowest Marks:", df["Marks"].min())
print("Total Students:", df["Name"].count())
print("Standard Deviation:", df["Marks"].std())
```
